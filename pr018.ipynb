```python
with open('input018.txt') as f:
    input_str = f.readlines()
```

```python
input_data = []
for line in input_str:
    input_data.append([int(n) for n in line.strip().split(' ')])
input_data
```

Though a brute force method is discouraged, let's try do it.

```python
path = [[75]]
level = 1
while level < 15:
    new_path = []
    for p in path:
        loc = input_data[level-1].index(p[-1])
        p_copy = p.copy()
        p_copy.append(input_data[level][loc])
        new_path.append(p_copy)
        
        p_copy = p.copy()
        p_copy.append(input_data[level][loc+1])
        new_path.append(p_copy)
    path = new_path
    level = len(new_path[0])
path
```

```python
max(sum(p) for p in path)
```
