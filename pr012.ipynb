# 12. [Highly divisible triangular number](https://projecteuler.net/problem=12)

The sequence of triangle numbers is generated by adding the natural numbers. So the $7^\text{th}$ triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

\begin{align}
1 & : 1 \\
3 & : 1, 3 \\
6 & : 1, 2, 3, 6 \\
10 & : 1, 2, 5, 10 \\
15 & : 1, 3, 5, 15 \\
21 & : 1, 3, 7, 21 \\
28 & : 1, 2, 4, 7, 14, 28
\end{align}

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
def triangle_number(n):
    '''return nth triangle number
    '''
    return sum(range(n+1))
```

Verify the function with the given examples.

```python
for i in range(1,7+1):
    print(i, triangle_number(i))
```

```python
import math as m

def factor(n):
    '''return factors of the given n
    '''
    fact = []
    for i in range(1, int(m.sqrt(n) + 2)):
        if n % i == 0:
            fact.extend((i, int(n / i)))
    return fact
```

```python
n = 1
t = 1
f = 1
while f < 500:
    n += 1
    t = triangle_number(n)
    f = len(factor(t))
print(n, t, f)
```
